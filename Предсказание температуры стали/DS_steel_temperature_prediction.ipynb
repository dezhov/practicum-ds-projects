# Оптимизация производственных расходов металлургического комбината: предсказание температуры стали на этапе обработки


**Постановка задачи**  
Чтобы оптимизировать производственные расходы, металлургический комбинат N решил уменьшить потребление электроэнергии на этапе обработки стали. Необходимо построить модель, которая предскажет температуру стали.

**Описание этапа обработки**  
Сталь обрабатывают в металлическом ковше вместимостью около 100 тонн. Чтобы ковш выдерживал высокие температуры, изнутри его облицовывают огнеупорным кирпичом. Расплавленную сталь заливают в ковш и подогревают до нужной температуры графитовыми электродами. Они установлены в крышке ковша.

Из сплава выводится сера (десульфурация), добавлением примесей корректируется химический состав и отбираются пробы. Сталь легируют — изменяют её состав — подавая куски сплава из бункера для сыпучих материалов или проволоку через специальный трайб-аппарат (англ. tribe, «масса»).

Перед тем как первый раз ввести легирующие добавки, измеряют температуру стали и производят её химический анализ. Потом температуру на несколько минут повышают, добавляют легирующие материалы и продувают сплав инертным газом. Затем его перемешивают и снова проводят измерения. Такой цикл повторяется до достижения целевого химического состава и оптимальной температуры плавки.

Тогда расплавленная сталь отправляется на доводку металла или поступает в машину непрерывной разливки. Оттуда готовый продукт выходит в виде заготовок-слябов (англ. slab, «плита»).


**Схема ковша-печи**



*1* – токоподвод; *2* – графитированные электроды; *3* – бункеры для ферросплавов; *4* – система удаления технологических газов; *5* – водоохлаждаемый свод (крышка); *6* – электрические дуги; *7* – рафинировочный шлак; *8* – жидкий металл; *9* – футерованный ковш; *10* – газовая продувка; *11* – устройство для перевозки ковша (сталевоз); *12* – продувочное устройство; *13* – выпускное отверстие с шиберным затвором.


***Цель исследования:***
* *Построение модели для определения температуры стали на этапе обработки со значением метрики MAE <= 8.7*

**Описание данных**  
Данные состоят из файлов, полученных из разных источников:

- *data_arc.csv* — данные об электродах;
- *data_bulk.csv* — данные о подаче сыпучих материалов (объём);
- *data_bulk_time.csv* — данные о подаче сыпучих материалов (время);
- *data_gas.csv* — данные о продувке сплава газом;
- *data_temp.csv* — результаты измерения температуры;
- *data_wire.csv* — данные о проволочных материалах (объём);
- *data_wire_time.csv* — данные о проволочных материалах (время).

Во всех файлах столбец `key` содержит номер партии.  
В файлах может быть несколько строк с одинаковым значением `key`: они соответствуют разным итерациям обработки.

```python
# напишем функцию для вывода информации о располагаемых данных
def print_info(df, key='key'):
    print('\nПервые объекты и признаки:')
    display(df.head(10))
    print('\nОбщая статиска:')
    display(df.describe().T)
    print('\nОбщая информация о датафрейме:\n')
    df.info()
    print(f'\nПропусков: {df.isna().sum().sum()}')
    print(f'Дубликатов: {df.duplicated().sum()}')
    print(f'Партий стали: {len(df[key].unique())}\n')
    
```

```python
# папишем функция для построения гистограммы
def plotting_hist(data, n_bins, title='', x_label='', y_label =''):
    fig = plt.figure(figsize=(20, 7))
    ax = fig.add_subplot()
    ax.hist(data, bins=n_bins, color='#8eb897', alpha=0.7)
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.grid(axis='y', alpha=0.3)
    ax.set_title(f'\n{title}\n', size=16)
    ax.set_xlabel(x_label, size=13)
    ax.set_ylabel(y_label, size=13)
    
    ax.axvline(data.median(),
               linestyle = ':',
               color = 'red',
               label = 'Медиана')
    ax.axvline(data.mean(),
               linestyle = ':',
               color = 'orange',
               label = 'Среднее')
    ax.axvline(data.quantile(0.1),
               linestyle = ':',
               color = 'black',
               label = '1% значений')
    ax.axvline(data.quantile(0.99),
               linestyle = ':',
               color = 'black',
               label = '99% значений')
    ax.legend(loc='best', fontsize='medium', shadow=True)

    plt.show()
    
```

```python
# приведём значения с датой в таблице 'df_arc' к типу 'datetime'
df_arc['Начало нагрева дугой'] = pd.to_datetime(df_arc['Начало нагрева дугой'])
df_arc['Конец нагрева дугой'] = pd.to_datetime(df_arc['Конец нагрева дугой'])

# проверим результат преобразования
df_arc.info()
```

```python
# создадим таблицу с новыми пизнаками на основе 'df_arc'
df_arc_info = df_arc.copy()

# расчитаем полную мощность
df_arc_info['Полная мощность'] = np.sqrt(df_arc_info['Активная мощность']**2
                                         +
                                         df_arc_info['Реактивная мощность']**2)
# расчитаем сообщённую энергию
df_arc_info['Время нагрева'] = (df_arc_info['Конец нагрева дугой']
                                -
                                df_arc_info['Начало нагрева дугой'])
df_arc_info['Время нагрева'] = df_arc_info['Время нагрева'].dt.seconds
df_arc_info['Энергия'] = (df_arc_info['Полная мощность']
                          *
                          df_arc_info['Время нагрева'])

# сгрупируем данные по партиям, просуммируем значения и расчитаем количество нагревов
df_arc_info['Нагревов'] = df_arc_info['key']
df_arc_info = (
    df_arc_info[['key', 'Полная мощность', 'Энергия', 'Нагревов']]
    .groupby('key')
    .agg({'Полная мощность': 'mean', 'Энергия': 'sum', 'Нагревов': 'size'})
    .rename(columns={'Полная мощность': 'full_power', 'Энергия': 'energy', 'Нагревов': 'heat_times'})
    .reset_index())

# выведим новыю таблицу
df_arc_info.head()
```

```python
# создадим таблицу с новыми признаками на основе 'df_temp'
df_temp_info = df_temp.copy()

# оставим партии, у которых проводилось больше одного замера температуры
df_temp_info = df_temp_info[df_temp_info.duplicated('key', keep=False) == True]

# сгрупируем данные по партиям, по первому и последнему замеру температуры
df_temp_info = (df_temp_info[['key', 'Температура']]
                .groupby(by = 'key')
                .agg(['first', 'last'])
                .reset_index())

# избавимся от мультииндекса и зададим наименованя признаков
df_temp_info.columns = df_temp_info.columns.droplevel()
df_temp_info.columns = ['key', 'first_temp', 'last_temp']

# выведим новыю таблицу
df_temp_info.head()
```
